```
#Joining the patterns data to the filtered places data

church_data = lds_church.join(
    patterns,
    lds_church.placekey == patterns.placekey,
    "inner"
)
```

# Data Wrangling

The next step would be to clean our data and pull out the variables we are looking for

```
# Extract the sampling rate from the dataset
church_data = church_data.withColumn("sampling_rate", church_data.normalized_visits_by_state_scaling / church_data.raw_visit_counts)

# find the data for Sunday visits
church_data = church_data.withColumn('visits_by_sunday', F.array_remove(
        F.transform('visits_by_day', lambda visits, i: F.when(
                i % 7 ==\
                    (F.dayofmonth(F.col('date_range_start')) - F.dayofweek(F.col('date_range_start')) + 7) % 7 # index of first Sunday
                , visits
            ).otherwise(-99999)
        ),
        -99999
        ))

# find average of sunday visits
query = """aggregate(
    `{col}`,
    CAST(0.0 AS double),
    (acc, x) -> acc + x,
    acc -> acc / size(`{col}`)
) AS  `avg_{col}`""".format(col="visits_by_sunday")
church_data = church_data.selectExpr("*", query)

church_data = church_data.withColumn("church_attendance", church_data.sampling_rate * church_data.avg_visits_by_sunday)
```

# Final Target

After aggregating tract data we are able to find the target

```
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

# based off church attendance find attendance per tract
church_agg = church_data.select(places.placekey, "date_range_start", F.explode("visitor_home_aggregation"), "church_attendance")
windowSpecAgg  = Window.partitionBy("placekey","date_range_start").orderBy("placekey")
church_agg = church_agg.withColumn("row",row_number().over(windowSpecAgg))\
  .withColumn("sum", F.sum(("value")).over(windowSpecAgg))

# final target of active church members per tract
church_agg = church_agg.withColumn("visitor_by_tract", church_agg.value / church_agg.sum * church_agg.church_attendance)\
    .groupBy("date_range_start", "key").agg(F.sum("visitor_by_tract").alias("target"))\
        .groupby("key").agg(F.mean("target").alias("target"))
church_agg.display()
```
